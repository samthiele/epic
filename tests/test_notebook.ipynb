```python
from epic import load_image, pick_single, pick_multi
```

```python
image = load_image('stinkbug.jpg')
image2 = image[::-1, :, : ].copy()
```

```python
P = pick_single( image, line=True, width=5,n=3, backend='nbAgg' )
```

```python
for p in P:
    print(p)
```

```python
P1, P2 = pick_multi( image, image2, line=True, width=5, n=3, backend='nbAgg' )
```

```python
for p1,p2 in zip(P1, P2):
    print (p1, ' -> ', p2 )
```
